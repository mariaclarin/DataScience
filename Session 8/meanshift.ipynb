```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift
```

```python
# Import dataset and converting the df to be in array form
# not using standardizing/normalizing bcs it is already in numeric form and can be comparable
df = pd.read_csv("HorrorSVMCookies.csv")
df = df[["Rating", "Raters"]]
arrayData = np.array(df)
x1 = arrayData[:,0]
y1 = arrayData[:,1]
plt.title("Rating vs Raters")
plt.scatter(x=x1, y=y1)
df.head()
```

```python
#meanshift algorithm
ms = MeanShift(bandwidth=None, seeds=None, bin_seeding = False, min_bin_freq=1, cluster_all=True, n_jobs=None, max_iter=300)
y_ms = ms.fit_predict(arrayData)

plt.scatter(arrayData[:, 0], arrayData[:, 1], c=y_ms)
ms.fit(arrayData)
labels = ms.labels_
```

```python
#the clustering and centroids based on the meanshift algorithm
cluster_centers = ms.cluster_centers_
n_clusters_ = len(np.unique(labels))
print("Estimated clusters:", n_clusters_)
colors = 10*['r.','g.','b.']
for i in range(len(arrayData)):
    plt.plot(arrayData[i][0], arrayData[i][1], colors[labels[i]], markersize = 10)
plt.scatter(cluster_centers[:,0],cluster_centers[:,1], marker="*",color='yellow',edgecolor='black', s=30, zorder=10)
plt.show()
```
