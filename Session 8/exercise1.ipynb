```python
#importing necessary libraries
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load data (I created the table in csv form)
df = pd.read_csv("exercise1.csv")
df = df[['x', 'y']]
df
```

```python
#k-means clustering
kmeans = KMeans(n_clusters=3, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=0).fit(df)
labels = kmeans.labels_

y_km = kmeans.fit_predict(df)
new_df = pd.DataFrame(data = df, columns = ['x', 'y'])
new_df['label_kmeans'] = labels
```

```python
#plotting the clusters and centroids
fig, ax = plt.subplots(figsize=(10, 10))
plt.scatter(new_df["x"][new_df["label_kmeans"] == 0], new_df["y"][new_df["label_kmeans"] == 0], 
            color = "blue", s=30, edgecolor='black',marker='s',label='cluster 1')
plt.scatter(new_df["x"][new_df["label_kmeans"] == 1], new_df["y"][new_df["label_kmeans"] == 1], 
            color = "red", s=30, edgecolor='black',marker='o',label='cluster 2')
plt.scatter(new_df["x"][new_df["label_kmeans"] == 2], new_df["y"][new_df["label_kmeans"] == 2], 
            color = "green", s=30, edgecolor='black',marker='v',label='cluster 3' )

#Using the determined centroid coordinates for C1, C2, and C3 based on the exercise
plt.scatter(1, 10, marker="o", s=50, color="yellow", edgecolor="black", label="centroids") #C(1,10)
plt.scatter(5, 8, marker="o", s=50, color="yellow", edgecolor="black") #C(5,8)
plt.scatter(9, 2, marker="o", s=50, color="yellow", edgecolor="black") #C(9,2)

ax.set_xlabel('x')
ax.set_ylabel('y')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
#plotting the clusters and centroids
fig, ax = plt.subplots(figsize=(10, 10))
plt.scatter(new_df["x"][new_df["label_kmeans"] == 0], new_df["y"][new_df["label_kmeans"] == 0], 
            color = "blue", s=30, edgecolor='black',marker='s',label='cluster 1')
plt.scatter(new_df["x"][new_df["label_kmeans"] == 1], new_df["y"][new_df["label_kmeans"] == 1], 
            color = "red", s=30, edgecolor='black',marker='o',label='cluster 2')
plt.scatter(new_df["x"][new_df["label_kmeans"] == 2], new_df["y"][new_df["label_kmeans"] == 2], 
            color = "green", s=30, edgecolor='black',marker='v',label='cluster 3' )

#The centroids have moved coordinates to be in its correct position as the mean of each cluster usinng kmeans algorithm
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='yellow', edgecolor='black', s=50,label='centroids')
ax.set_xlabel('x')
ax.set_ylabel('y')


plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```
